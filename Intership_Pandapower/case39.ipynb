```python
import pandapower as pp #import pandapower
import numpy as np
import pandapower.networks as pn 
from pandapower.estimation import estimate
import pandapower.plotting as plot
%matplotlib inline
import matplotlib.pyplot as plt
```

## IEEE 39-Bus System
The “IEEE 39 bus system is well known as 10-machine New-England Power System. Generator 1 represents the aggregation of a large number of generators.
<img src="pics/IEEE39.png" width="600">

Calls the pickle file case39.p which data origin is PYPOWER. This network was published the first time in
G. Bills et al., On-line stability analysis study, RP 90-1, E. P. R. I. North American Rockwell Corporation,
Edison Electric Institute, Ed. IEEE Press, Oct. 1970,. Some more information about this network are
given by Illinois University case 39. Because the Pypower data origin proposes vn_kv=345 for all nodes
the transformers connect node of the same voltage level.

```python
net = pn.case39()
```

```python
net
```

```python
net.res_bus
```

```python
net_val = net.res_bus.values
net_index = net.res_bus.index.values
```

```python
net_val.shape, net_index.shape
```

```python
net_arr = np.c_[net_index,net_val]
```

```python
net_arr.shape
```

```python
sigma_vm = 0.1
sigma_p = 20
sigma_q = 20
measurment_index = net_arr[:,0].astype(int)
measurment_vm = net_arr[:,1] + np.random.normal(0,sigma_vm,net_arr.shape[0])
measurment_p = -net_arr[:,3] + np.random.normal(0,sigma_p,net_arr.shape[0])
measurment_q = -net_arr[:,4] + np.random.normal(0,sigma_q,net_arr.shape[0])
```

```python
measurment_index, measurment_vm,measurment_p,measurment_q
```

```python
measurment_vm[0]
```

```python
for i in range(net_arr.shape[0]):
    pp.create_measurement(net, "v", "bus", measurment_vm[i], sigma_vm, measurment_index[i])
    pp.create_measurement(net, "p", "bus", measurment_p[i], sigma_p, measurment_index[i])        
    pp.create_measurement(net, "q", "bus", measurment_q[i], sigma_q, measurment_index[i]) 
net.measurement
```

```python
success = estimate(net, init='flat', tolerance=1e-06, maximum_iterations=100,ref_power=100000000)
print(success)
```

```python
colors = ["b", "g", "r", "c", "y"]
lc = plot.create_line_collection(net, net.line.index, color=colors[0], zorder=1) #create lines
bc = plot.create_bus_collection(net, net.bus.index, size=0.1, color=colors[0], zorder=2) #create buses
plot.draw_collections([lc,bc], figsize=(8,6)) # plot lines and buses
```

```python
buses = net.bus.index.tolist() # list of all bus indices
coords = zip(net.bus_geodata.x.loc[buses].values+0.1, net.bus_geodata.y.loc[buses].values+0.1) # tuples of all bus coords

bic = plot.create_annotation_collection(size=0.2, texts=np.char.mod('%d', buses), coords=coords, zorder=3, color=colors[1])
plot.draw_collections([lc, bc, bic], figsize=(80,60)) # plot lines, buses and bus indices
```
